## Visualization & Explanatory Analysis

Pada tahap ini dilakukan visualisasi data setelah mendapatkan jawaban dari permasalahan tersebut

### Pertanyaan 1:Bagaimana performa penjualan dan revenue perusahaan dalam beberapa bulan terakhir?

```python

monthly_orders_df = orders_order_items_df.groupby('month').agg({
    'order_id': 'nunique',
    'price': 'sum',
}).reset_index()
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    'order_id': 'total_orders',
    'price': 'price'
}, inplace=True)

# Menampilkan hasil akhir
print(monthly_orders_df)

```

```python
monthly_orders_df = orders_order_items_df.groupby('month').agg({
    'order_id': 'nunique',
    'price': 'sum',
}).reset_index()
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    'order_id': 'total_orders',
    'price': 'price'
}, inplace=True)

plt.figure(figsize=(10, 5))
plt.plot(monthly_orders_df["month"], monthly_orders_df["total_orders"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Number of Orders per Month (2021)", loc="center", fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 2: Bagaimana demografi pelanggan yang kita miliki?

```python
bystate_df = customers_df.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
bystate_df
plt.figure(figsize=(10, 5))
colors_ = ["#72BCD4"] + ["#D3D3D3"] * (len(bystate_df) - 1)
sns.barplot(
    x="customer_count",
    y="customer_state",
    data=bystate_df.sort_values(by="customer_count", ascending=False),
    palette=colors_,
    hue=None,
    legend=False
)
plt.title("Number of Customer by States", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

Oleh karena itu, penting bagi perusahaan untuk menganalisis lebih dalam faktor-faktor yang memengaruhi penurunan penjualan di bulan-bulan tertentu dan mengoptimalkan strategi pemasaran untuk memanfaatkan periode penjualan yang menguntungkan.
